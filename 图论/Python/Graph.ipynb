- 图论的最基本操作

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图 G
G = nx.Graph()
# 创建一个无向图 NG
NG = nx.DiGraph()
NG.clear()
G.clear()

# 添加一个边 add_edge(),删除一条边 remove_edge(),添加节点和删除节点同理
G.add_node(0, color='red')
G.add_edges_from([(0, 1), (1, 2), (2, 3)], weight=0.8)
G.remove_edge(0, 1)
# 这里的第三个参数可以有许多，权重，流量
G.add_edge(0, 1, weight = 0.8)

nx.draw(G)
plt.show()
```

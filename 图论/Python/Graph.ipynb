- 图论的最基本操作

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图 G
G = nx.Graph()
# 创建一个无向图 NG
NG = nx.DiGraph()
NG.clear()
G.clear()

# 添加一个边 add_edge(),删除一条边 remove_edge(),添加节点和删除节点同理
G.add_node(0, color='red')
G.add_edges_from([(0, 1), (1, 2), (2, 3)], weight=0.8)
G.remove_edge(0, 1)
# 这里的第三个参数可以有许多，权重，流量
G.add_edge(0, 1, weight = 0.8)

nx.draw(G)
plt.show()
```

对于无向图(有向图)，遍历边和节点：

```python
# 遍历节点
node = [n for n in G]  # 等价于 list(G)
# 如果想要节点的属性，就调用 G.nodes(data=True)
# 输出：NodeDataView({0: {'color': 'red'}, 1: {}, 2: {}, 3: {}})

# 遍历边：和 C++ 类似
edge = [e for e in G[0]] # 以 0 为节点的边
print(edge, G[0])
# 遍历全部的边以及属性 G.edges()
```

单源最短路的算法：
   * 非负权重：dijkstra 算法
* 负权重：bellman-ford 算法

```python
e = [(1, 2, {"weight" : 2}), (2, 3, {"weight" : 2}), (2, 4, {"weight" : 1}), (1, 3, {"weight" : 5}), (3, 4, {"weight" : 3}), (1, 4, {"weight" : 4})]

NG.add_edges_from(e)

print(nx.shortest_path(NG, source=1, weight="weight"))
print(nx.shortest_path_length(NG,source=1, weight="weight"))
```
